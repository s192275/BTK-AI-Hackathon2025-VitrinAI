```python
# 3. Sesli komut ile ürün açıklaması oluşturma -> OpenAI Whisper turbo modelini kullanıyoruz.
import sounddevice as sd
import numpy as np
import time
from faster_whisper import WhisperModel
model_size = "turbo"
model = WhisperModel(model_size, device="cpu")  # CT2 formatta model
fs = 16000
chunk_duration = 9     # 2 saniyelik buffer
total_duration = 10     # toplam 10 saniye dinleme
buffer = np.zeros(0, dtype=np.float32)

def callback(indata, frames, time, status):
    global buffer
    buffer = np.concatenate((buffer, indata[:, 0]))

with sd.InputStream(callback=callback, channels=1, samplerate=fs):
    print(f"Dinleniyor... Toplam {total_duration}s, {chunk_duration}s'lik parçalar halinde.")
    start = time.time()
    while True:
        elapsed = time.time() - start
        if elapsed >= total_duration:
            break

        if len(buffer) >= fs * chunk_duration:
            chunk = buffer[:fs * chunk_duration]
            buffer = buffer[int(fs * (chunk_duration - 0.5)):]  # overlap: 0.5s
            segments, info = model.transcribe(chunk, language="tr")
            text = " ".join([seg.text for seg in segments])
            print(text)

        time.sleep(0.1)  # CPU yükünü düşür

    print("⏹ Süre doldu, program bitiriliyor.")
```

```python
from transformers import Wav2Vec2Processor, Wav2Vec2ForCTC, pipeline
import sounddevice as sd
import numpy as np
import torch
duration = 10
sample_rate = 16000
processor = Wav2Vec2Processor.from_pretrained("Sercan/wav2vec2-xls-r-300m-tr")
stt_model = Wav2Vec2ForCTC.from_pretrained("Sercan/wav2vec2-xls-r-300m-tr")
print("10 saniyelik ses kaydı başlıyor, lütfen konuşun...")
audio = sd.rec(int(duration * sample_rate), samplerate=sample_rate, channels=1, dtype='float32')
sd.wait()
audio = np.squeeze(audio)

input_values = processor(audio, sampling_rate=sample_rate, return_tensors="pt", padding=True).input_values
with torch.no_grad():
    logits = stt_model(input_values).logits
    predicted_ids = torch.argmax(logits, dim=-1)
    transcription = processor.batch_decode(predicted_ids)[0]
print(transcription)
```

```python
from ddgs import DDGS
import requests
from bs4 import BeautifulSoup
import re

headers = {"User-Agent": "Mozilla/5.0"}

def extract_ingredients(url):
    try:
        response = requests.get(url, headers=headers, timeout=10)
        soup = BeautifulSoup(response.text, "html.parser")
        text = soup.get_text(separator="\n").lower()
        match = re.search(r"(içindekiler|içeriği)\s*[:：]?\s*(.+?)(\n|$)", text)
        if match:
            return match.group(2).strip()
        return "İçindekiler bulunamadı"
    except Exception as e:
        return f"Hata: {e}"

product = "ülker çikolatalı gofret"
with DDGS() as ddgs:
    results = ddgs.text(f"{product} içindekiler site:getir.com", max_results=3)

for item in results:
    print(f"\n🔎 {item['title']}\nURL: {item['href']}")
    ingredients = extract_ingredients(item['href'])
    print(f"📦 İçindekiler: {ingredients}")
```
